```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python


def simular_embarque():
    asientos = [None] * 100
    disponibles = list(range(100))


    asiento_elegido = random.choice(disponibles)
    asientos[asiento_elegido] = 1
    disponibles.remove(asiento_elegido)


    for pasajero in range(2, 101):
        asiento_asignado = pasajero - 1
        if asiento_asignado in disponibles:
            asientos[asiento_asignado] = pasajero
            disponibles.remove(asiento_asignado)
        else:
            asiento_aleatorio = random.choice(disponibles)
            asientos[asiento_aleatorio] = pasajero
            disponibles.remove(asiento_aleatorio)


    return asientos[99] == 100, asientos[99]


simulaciones = 1_000_000
exitos = 0
asientos_finales = []

for _ in range(simulaciones):
    exito, asiento_100 = simular_embarque()
    if exito:
        exitos += 1
    asientos_finales.append(asiento_100)


probabilidad = exitos / simulaciones
print(f"Probabilidad de que el pasajero 100 se siente en su asiento: {probabilidad:.4f}")


plt.hist(asientos_finales, bins=range(1, 102), edgecolor='black', align='left')
plt.title("Asiento pasajero 100")
plt.xlabel("NÃºmero de asiento")
plt.ylabel("Frecuencia")
plt.show()
```
